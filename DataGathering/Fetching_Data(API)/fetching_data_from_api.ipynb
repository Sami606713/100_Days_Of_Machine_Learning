```python
import requests
import pandas as pd
import numpy as np
```

# Get the data from new api

1. **web_url**: https://newsapi.org/ 
2. **Api**: https://newsapi.org/v2/everything?q=tesla&from=2024-03-16&sortBy=publishedAt&apiKey=be805478ab5840a3b820a974b56fe770
3. **Jon viewer:[Paste the api conetent in this website to see the clear content]** https://jsonviewer.stack.hu/

```python
url='https://newsapi.org/v2/everything?q=tesla&from=2024-03-16&sortBy=publishedAt&apiKey=be805478ab5840a3b820a974b56fe770'
try:
    response=requests.get(url)
except Exception as e:
    print(e)
```

```python
response.json()
```

**We can see that our necessary data are present in the `articles` field so we can get only `articles`** 

```python
final=response.json()['articles']
final
```

```python
author=[]
title=[]
desc=[]
content=[]
publish_date=[]
url_to_image=[]

for i in final:
    author.append(i['author'])
    title.append(i['title'])
    desc.append(i['description'])
    content.append(i['content'])
    publish_date.append(i['publishedAt'])
    url_to_image.append(i['urlToImage'])
```

# Now convert this into dataframe

```python
dic={
    "title":title,
    "desc":desc,
    "content":content,
    "date":publish_date,
    "url_to_image":url_to_image,
    "author":author
}
dic
```

# Final Dataframe

```python
df=pd.DataFrame(dic)
df.sample(10)
```
