# Load the lobraries and dataset

```python
import pandas as pd
import numpy as np
```

# Read_csv

```python
salaries=pd.read_csv('salaries.csv')
salaries.head(2)
```

# use_col
1. **It is use to get the specific columns**

```python
pd.read_csv('salaries.csv',usecols=['work_year','job_title','salary_in_usd'])
```

# skiprows
1. **It is use to skip the any rows**

```python
pd.read_csv('salaries.csv',skiprows=[0,1])
```

**Now skip the even rows**

```python
pd.read_csv('salaries.csv',skiprows=[i for i in range(10) if i%2==0])
```

# Sep keyword
1. It is used to saperate each column with specific character

```python
pd.read_csv('salaries.csv',sep=",")
```

# index_col
1. It is used to set any column as index

```python
pd.read_csv('salaries.csv',index_col='work_year')
```

# skip_blank_lines
1. It is used to skip the bad lines

```python
pd.read_csv('salaries.csv',skip_blank_lines=False)
```

# chunksize
1. It is used to load the data in the form of chunk

```python
chunk_df=pd.read_csv('salaries.csv',chunksize=1000)
for i in chunk_df:
    print(i.shape)
```

# nrows
1. It is used to load limited rows

```python
pd.read_csv('salaries.csv',nrows=500)
```

# header
1. It is used to set any rows as header

```python
pd.read_csv('salaries.csv',header=1)
```

# encoding=None
1. It is used if get the `encoding error` some encoding parameter are `utf-8 latin-1`

```python
pd.read_csv('salaries.csv',encoding='utf-8')
```

# NA-Values
1. It is used to treat any value as NA

```python
pd.read_csv('salaries.csv',na_values='Data Scientist')
```
