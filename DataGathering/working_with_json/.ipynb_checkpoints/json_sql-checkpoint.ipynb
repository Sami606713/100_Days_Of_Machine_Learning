# working with sql

```python
import pandas as pd
import requests
import numpy as np
import pymysql
import warnings
warnings.filterwarnings('ignore')
```

# Make the connection of database

```python
connection=pymysql.connect(host="127.0.0.1",
                user='root',
                password='root',
               database="case_studies")
connection
```

# read the data from database

```python
df=pd.read_sql_query("""select * from salaries""",con=connection)
df.head(2)
```

**We can apply all the sql queries**

```python
df
```

# read the data in the form of chunk

```python
chunk_df=pd.read_sql_query("""select * from salaries""",con=connection,chunksize=5000,)
```

```python
for i in chunk_df:
    print(i.shape)
```
