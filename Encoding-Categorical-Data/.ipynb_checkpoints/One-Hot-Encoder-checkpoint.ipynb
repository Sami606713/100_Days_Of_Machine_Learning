# One-Hot-Encoder

# Import Packages

```python
import pandas as pd
import numpy as np
```

# Load Data-Set

```python
df=pd.read_csv('onlinefoods.csv')
df.head(2)
```

```python
df.drop(columns='Unnamed: 12',inplace=True)
```

```python
df.shape
```

```python
df.head()
```

# Encode using pandas

```python
pd.get_dummies(df,columns=['Gender','Marital Status','Occupation','Output'],drop_first=True,sparse=False)
```

# ordinal-col
1. `Educational Qualifications`
# Nominal-col
1. `Gender`, `Marital Status`, `Occupation`, `Output`

# label encoder
1. Encode the target col `Feedback`

```python
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder,OneHotEncoder
```

```python
encode=LabelEncoder()
df['Feedback']=encode.fit_transform(df['Feedback'])
```

# Processing
1. First split the data into test and train set
2. Encode the ordinal column
3. Encode the nominal column

# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='Feedback'),df['Feedback'],test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

# Encode the Oridnal Col `Educational Qualifications`

```python
odr=OrdinalEncoder(categories=[['Uneducated','School','Post Graduate','Graduate','Ph.D']])
x_train['Educational Qualifications']=odr.fit_transform(x_train[['Educational Qualifications']])
x_test['Educational Qualifications']=odr.transform(x_test[['Educational Qualifications']])
```

# Encode the `Nominal Col`

```python
ohe=OneHotEncoder(drop='first',sparse=False,dtype=np.int32,handle_unknown='ignore')
x_train_encode=ohe.fit_transform(x_train[['Gender','Marital Status','Occupation','Output']])
x_test_encode=ohe.transform(x_test[['Gender','Marital Status','Occupation','Output']])
```

```python
x_train_encode.shape
```

# Combine Encode and Non-Encode columns

```python
new_x_train=np.hstack((x_train.drop(columns=['Gender','Marital Status','Occupation','Output']),x_train_encode))
new_x_train
```

```python
new_x_test=np.hstack((x_test.drop(columns=['Gender','Marital Status','Occupation','Output']),x_test_encode))
new_x_test
```
