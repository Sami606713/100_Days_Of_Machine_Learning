# Function Transformer
1. `Function transfomer` is are used to apply the transformation on numerical col so that the given column should be normally distrubuted
    1. `Log Transformation` is used fro left skewed data.
    2. `sqrt transformation` data is used for right skewed data.

# import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
```

# load dataset

**we can used only 3 numerical col and output col**

```python
df=pd.read_csv('onlinefoods.csv',usecols=['Age','latitude','longitude','Feedback'])
df.head(2)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

# Check the distrubution

```python
sns.kdeplot(df['longitude'])
```

# Saperate feature and label

```python
feature=df.drop(columns=['Feedback'])
label=df['Feedback']
```

# Encode the output col

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode=LabelEncoder()
label=encode.fit_transform(label)
```

# Split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,label,test_size=_size=0.2,random_state=42)
```

# Train  the model without `Function transformation`

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pre=model.predict(x_test)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pre)
```

**we got 85 accuracy now apply function transformer and see the result**

# Apply `Function Transformer`

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
trans=FunctionTransformer(np.log1p)
x_train_tran=trans.fit_transform(x_train)
x_test_tran=trans.transform(x_test)
```

# Fit the model again

```python
model=LogisticRegression()
```

```python
model.fit(x_train_tran,y_train)
```

```python
pre=model.predict(x_test_tran)
```

```python
accuracy_score(y_test,pre)
```

# Function for see the result of `transformation`

```python
def transformation(transformation_name):
    x_train, x_test, y_train, y_test = train_test_split(feature, label, test_size=0.2, random_state=42)
    
    trans = FunctionTransformer(transformation_name)
    x_train_tran = trans.fit_transform(x_train)
    x_test_tran = trans.transform(x_test)
    
    model = LogisticRegression()
    model.fit(x_train_tran, y_train)
    pre = model.predict(x_test_tran)
    print("Accuracy:", accuracy_score(y_test, pre))
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
    ax1.set_title("Before transformation kde plot")
    sns.kdeplot(x_train['Age'], ax=ax1)
    
    ax2.set_title("Before transformation QQ plot")
    stats.probplot(x_train['Age'], dist="norm", plot=ax2)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
    ax1.set_title("After transformation kde plot")
    sns.kdeplot(x_train_tran['Age'], ax=ax1)
    
    ax2.set_title("after transformation QQ plot")
    stats.probplot(x_train_tran['Age'], dist="norm", plot=ax2)
    
    plt.show()
```

```python
transformation(lambda x:x+x/x)
```
