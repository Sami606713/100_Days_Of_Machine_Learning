# Imbalance Data
1. Imbalance data is a data in which one class dominate the other classes.
#### Example
1. In a dataset of email spam classifier, there may be only 30 Spam for every 2000 Not spam emails. 

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Load the image
img = mpimg.imread('imbalance-data.png')
plt.figure(figsize=(15,5))
# Display the image
plt.imshow(img)
plt.axis('off')  # Turn off axis
plt.show()
```
