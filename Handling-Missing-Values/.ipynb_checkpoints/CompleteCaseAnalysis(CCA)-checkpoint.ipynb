# Complete Case Analysis
1. **In Complete case analysis we can remove all these rows that contain missing values.**
2. **We can apply CCA if we are sure if data is missing completely at ranmdom (MCAR).**
3. **Generally we can remove value is value is missiing less then 5%.**

**Notice**
1. **Before removing and after removing distrubution of data should be same or almost same**

# Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Load Dataset

```python
df=pd.read_csv('data_science_job.csv')
df.head()
```

# Data Preprocessing

```python
df.shape
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()/len(df)*100
```

# Conclussion
1. We can see that see that `city_development_index` contain `2%`, `gender` contain `23%`, `enrolled_university` contain `2%`,  `education_level` contain `2%`, `major_discipline` contain `14%`, `company_size` contain `30%`, `company_type` contain `32%`, `training_hours` contain `3%` missing valules

**Now we can get the column that conatin less then or equal to `5%` missing values**

```python
missing_less_5=[col for col in df.columns if (df[col].isnull().sum()/len(df)*100)<=5 and (df[col].isnull().sum()/len(df)*100)>0]
missing_less_5
```

# Now drop the rows form these col and check the distrubtion

```python
df[missing_less_5].isnull().sum()/len(df)*100
```

```python
temp=df[missing_less_5].dropna()
```

```python
temp.head(2)
```

# Now check the distrubution

# city_development_index

```python
plt.figure(figsize=(15,7))
plt.title("Distrubution of city_development_index before and after removing")
sns.histplot(df['city_development_index'],color='red')
sns.histplot(temp['city_development_index'],color='green')
plt.show()
```

# experience

```python
plt.figure(figsize=(15,7))
plt.title("Distrubution of experience before and after removing")
sns.histplot(df['experience'],color='red')
sns.histplot(temp['experience'],color='green')
plt.show()
```

# training_hours

```python
plt.figure(figsize=(15,7))
plt.title("Distrubution of training_hours before and after removing")
sns.histplot(df['training_hours'],color='red')
sns.histplot(temp['training_hours'],color='green')
plt.show()
```

# Conclussion
1. We can see that there is no mightly change except `experience` data is almost balance the distrubution
