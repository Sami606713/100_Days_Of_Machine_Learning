## IQR 
1. **The Interquartile Range (IQR) technique identifies outliers by calculating the range between the first and third quartiles of a dataset. It's robust against skewed data due to its reliance on quartiles rather than the mean and standard deviation.**
## Assumption
1. **They are working on skewed data.**

# Import packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings('ignore')
```

# Load Data

```python
df=pd.read_csv('order_details.csv')
df.head(2)
```

```python
df.shape
```

**we have `1500` rows and `5` columns**

```python
df.isnull().sum()
```

# Check the destribution

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))

ax1.set_title("Distribution of amount")
sns.kdeplot(df['amount'], ax=ax1)

ax2.set_title("Boxplot of amount")
sns.boxplot(df['amount'], ax=ax2)

ax3.set_title("Distribution of profit")
sns.kdeplot(df['profit'], ax=ax3)

ax4.set_title("Boxplot of profit")
sns.boxplot(df['profit'], ax=ax4)

plt.show()
```

# Conclussion
1. **we can see that `amount` col is skewed now we can apply the `IQR` technique.**

1. `min_value = q1 - 1.5 * IQR`
2. `max_value = q1 + 1.5 * IQR`
3. `IQR = q3-q1`
4. `q1 => 25 percentile, q2=> 50 percentile, q3=> 75 percentile`

```python
def iqr_technique(col):
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    
    iqr=q3-q1
    
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    
    dic={
        "upper":upper_bound,
        'lower':lower_bound
        }
    return dic        
```

# Now apply the function

```python
result=iqr_technique("amount")
result
```

**These are the upper or lower limit now apply the technique `Trim` or `Cap`**

# Triming
1. In trimming we can simply remove them.

```python
df[(df['amount']>result['upper']) | (df['amount']<result['lower'])]
```

**These are the outlier rows**

```python
new_df=df[~((df['amount']>result['upper']) | (df['amount']<result['lower']))]
new_df
```

# Now Check the dustribution after `trimming`

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

ax1.set_title("Distribution of amount before apply IQR")
sns.kdeplot(df['amount'], ax=ax1)

ax2.set_title("Distribution of amount after apply IQR")
sns.kdeplot(new_df['amount'], ax=ax2)

plt.show()
```

```python
plt.title('BoxPlot after applying IQR')
new_df['amount'].plot(kind='box')
plt.show()
```

**we can see that we can remove reduce some outlier after `Trimming`**

# Apply `Capping`

```python
df['amount_cap']=np.where(df['amount']>result['upper'],
        result['upper'],
        np.where(df['amount']<result['lower'],
                result['lower'],
                df['amount']))
```

```python
df.head(2)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 6))

ax1.set_title("Distribution of amount before apply IQR")
sns.kdeplot(df['amount'], ax=ax1)

ax2.set_title("Distribution of amount after apply IQR")
sns.kdeplot(df['amount_cap'], ax=ax2)

plt.show()
```

```python
plt.title('BoxPlot after applying IQR')
df['amount_cap'].plot(kind='box')
plt.show()
```
