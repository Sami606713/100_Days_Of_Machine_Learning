# Winserization or percentile method
1. **This is a technique for handling outliers. In this technique we can use quantile like `90` percentile or `10 percentile` etc we can decide percentile according the the best result**
2. **They are applied on skewed data.**

# Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# Load Data

```python
df=pd.read_csv("order_details.csv")
```

# Check the dustribution

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
ax1.set_title('Distrubution of amount')
sns.kdeplot(df['amount'],ax=ax1)

ax2.set_title('Distrubution of profit')
sns.kdeplot(df['profit'],ax=ax2)
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
ax1.set_title('Distrubution of amount')
sns.boxplot(df['amount'],ax=ax1)

ax2.set_title('Distrubution of profit')
sns.boxplot(df['profit'],ax=ax2)
plt.show()
```

# Conclussion
1. **we can see that `amount` is skewed now we can apply `percentile` approach on `amount` col.** 

```python
def find_bounds(col,per1,per2):
    lower_bound=df[col].quantile(per1)
    upper_bound=df[col].quantile(per2)
    
    dic={
        "upper_bound":upper_bound,
        "lower_bound":lower_bound
    }
    return dic
```

```python
result = find_bounds('amount', 0.10, 0.90)
result
```

**These are the upper and lower bound now apply `Trimming` and `Capping`.**

```python
df['amount'].quantile(0.99)
```

# Trimming

```python
df[((df['amount']>=result['upper_bound'])|(df['amount']<=result['lower_bound']))].shape
```

# conclussion
1. These are the outlier values or rows.

```python
new_df=df[~((df['amount']>=result['upper_bound'])|(df['amount']<=result['lower_bound']))]
new_df.shape
```

# Now check the distrubution again

```python
new_df['amount'].plot(kind='box')
```

```python
new_df['amount'].plot(kind='kde')
```

# Conclussion
1. **See that we can remove some outliers**

# Capping

```python
df["amount_cap"]=np.where(df['amount']>result['upper_bound'],
        result['upper_bound'],
        np.where(df['amount']<=result['lower_bound'],
                result['lower_bound'],df['amount']))
```

```python
df.head(2)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
ax1.set_title('before applying winzsrization')
sns.boxplot(df['amount'],ax=ax1)

ax2.set_title('after applying winzsrization')
sns.boxplot(df['amount_cap'],ax=ax2)
plt.show()
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
ax1.set_title('before applying winzsrization')
sns.kdeplot(df['amount'],ax=ax1)

ax2.set_title('after applying winzsrization')
sns.kdeplot(df['amount_cap'],ax=ax2)
plt.show()
```

**see that we can remove outliers almost by aplying `capping` and `trimming`.**
