# LinearRegression
- In this we will apply the linear regression on build-in houses dataset

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset
california_housing = fetch_california_housing(as_frame=True)
```

```python
california_housing
```

```python
# Convert to a DataFrame
data = pd.concat([california_housing.data, california_housing.target], axis=1)

# Rename the target column for clarity
data.rename(columns={'MedHouseVal': 'Target'}, inplace=True)
```

```python
data.head()
```

```python
sns.pairplot(data)
```

# Saperate Feature and label

```python
feature=data.drop(columns='Target')
label=data['Target']
```

# Train test split

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
x_train,x_test,y_train,y_test=train_test_split(feature,label,test_size=0.2,random_state=42)
```

```python
col_pipe=Pipeline(steps=[
    ("scale",StandardScaler())
])
```

```python
process=ColumnTransformer(transformers=[
    ("process",col_pipe,feature.columns)
],remainder="passthrough")
process
```

```python
x_train_scale=process.fit_transform(x_train)
x_test_scale=process.transform(x_test)
```

```python
x_train_scale
```

```python
x_test_scale
```

# Apply Model
- `y = mx + b`

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
```

```python
model=LinearRegression()
# model=Ridge(alpha=0.01)
model=Lasso(alpha=0.02)
model.fit(x_train_scale,y_train)
```

```python
b=model.intercept_
b
```

```python
m=model.coef_
m
```

# Prediction

```python
pre=model.predict(x_test_scale)
pre
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_absolute_error(y_test,pre)
```

```python
mean_squared_error(y_test,pre)
```

```python
r2_score(y_test,pre)
```

```python
fig,ax=plt.subplots(figsize=(15,7))
sns.regplot(x=y_test,y=pre,color='red',)
```

# Hyperparameter Tunning

```python
models={
    "LinearRegrssion":LinearRegression(),
    "Ridge":Ridge(),
    "Lasso":Lasso()
}
```

### Parameters

```python
param_grid={
    "LinearRegrssion":{},
    "Ridge":{
        "alpha":[0.1, 1.0, 10.0, 100.0]
    },
    "Lasso":{
        "alpha":[0.1, 1.0, 10.0, 100.0]
    }
}
```

```python
best_estimator={}
for model_name,model in models.items():
    model_param=param_grid[model_name]
    print(f"model------>{model_name} |  parameter------->{model_param}")
    
    grid_search=GridSearchCV(model,model_param,cv=5,scoring='r2',n_jobs=-1)
    
    grid_search.fit(x_train_scale,y_train)
    best_estimator[model_name]=grid_search.best_estimator_
    
    print(f"Best parameters of {model_name} is {grid_search.best_estimator_}")
    print(f"Best score of {model_name} is {grid_search.best_score_}")
    print("*"*50)
    print()
```

```python
for model_name ,estimator in best_estimator.items():
    score=estimator.score(x_test_scale,y_test)
    print(f"Test score for {model_name}: {score}")
```
