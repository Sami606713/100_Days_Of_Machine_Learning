# Load Dataset

```python
import numpy as np
import pandas as pd

# Generating CGPA between 2.0 and 4.0
cgpa = np.random.uniform(2.0, 4.0, 1000)

# Generating Placement Salary based on a combination of linear and non-linear functions with some randomness
# Salary = m1 * CGPA + m2 * (CGPA)^2 + m3 * exp(CGPA) + c + random_error
m1 = 5000
m2 = -5000
m3 = 2000
c = 20000
random_error = np.random.normal(0, 3000, 1000)  # Introducing some randomness
salary = m1 * cgpa + m2 * cgpa**2 + m3 * np.exp(cgpa) + c + random_error
# Creating a DataFrame
data = pd.DataFrame({'CGPA': cgpa, 'Placement Salary': salary})

data
```

# Conclussion
- In this we can use to predict the `cgpa` based on `iq`

# Check the relationship

```python
sns.scatterplot(data=data,x='CGPA',y='Placement Salary')
```

# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x=data['CGPA']
y=data['Placement Salary']
```

```python
x.head(4)
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_train_reshape=x_train.values.reshape(-1,1)
```

```python
y_train_reshape=y_train.values.reshape(-1,1)
```

# Build Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(x_train_reshape,y_train_reshape)
```

# Prediction

```python
x_test_reshape=x_test.values.reshape(-1,1)
```

```python
pre=model.predict(x_test_reshape)
pre
```

# Check `m` and `b`

```python
b=model.intercept_
b
```

```python
m=model.coef_
m
```

# Now check the best fit line that model find

```python
plt.figure(figsize=(15,7))
sns.scatterplot(data=data,x='CGPA',y='Placement Salary')
plt.plot(x_test_reshape,pre,color='red',label='Best Fit line')
plt.legend()
plt.show()
```

# Conclussion
- In summary linear regression find the best fit line so that they can pass closly to every point.
