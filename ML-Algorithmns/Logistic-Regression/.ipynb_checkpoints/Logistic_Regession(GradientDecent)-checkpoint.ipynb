```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

# Generate Dataset

```python
# Generate the dataset
X, y = make_classification(n_samples=100,   # Number of samples
                           n_features=2,     # Number of features (inputs)
                           n_informative=1,  # Number of informative features
                           n_redundant=0,    # Number of redundant features
                           n_classes=2,      # Number of classes (output)
                           n_clusters_per_class=1,
                           hypercube=False,class_sep=5,
                           random_state=42)

# Create a DataFrame with the input features and output
df = pd.DataFrame(X, columns=['feature1', 'feature2'])
df['output'] = y
```

```python
df.head()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(df['feature1'],df['feature2'],c=df['output'])
plt.show()
```

```python
def sigmoid(z):
    return (1/(1+np.exp(z)))
```

```python
def gradient(X,y,lr=0.01,epocs=1):
    X=np.insert(X,0,1,axis=1)
    
    weights=np.ones(X.shape[1])
    
    for i in range(epocs):
        y_hat=sigmoid(np.dot(X,weights))
        error=(y-y_hat)
        
        gradient=np.dot(X.T,error)/X.shape[0]
        
        weights+=lr*gradient
    
    return weights[0],weights[1:]
```

```python
intercept,coffi=gradient(X,y,epocs=10000)
```

```python
intercept
```

```python
coffi
```

```python
m=-(coffi[0]/coffi[1])
b=-(intercept/coffi[1])
```

```python
print(m)
print(b)
```

```python
x_input=np.linspace(-3,3,2000)
y_input=m*x_input + b
```

```python
plt.figure(figsize=(15,7))
plt.plot(x_input,y_input,)
plt.scatter(X[:,0],X[:,1],c='Red')
plt.show()
```

# Apply Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model.fit(X,y)
```

```python
lr_coffi=model.coef_
lr_intercept=model.intercept_
```

```python
m2 = -(lr_coffi[0][0] / lr_coffi[0][1])
b2 = -(lr_intercept / lr_coffi[0][1])
```

```python
print(m2,b2)
```

```python
x_input2=np.linspace(-3,3,100)
y_input2=m2*x_input2 + b2
```

```python
y_input2
```

```python
plt.figure(figsize=(15,6))
plt.plot(x_input2,y_input2,c='Blue',label='Logistic Regression')
plt.scatter(X[:,0],X[:,1])
plt.legend()
plt.show()
```
