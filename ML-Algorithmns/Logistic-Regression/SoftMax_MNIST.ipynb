```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("Digit_classifier.csv")
```

```python
df.sample(10)
```

# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns='label'),df['label'],test_size=0.2,random_state=43)
```

```python
x_train
```

# Apply PCA

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
x_train_transform=pca.fit_transform(x_train)
x_test_transfom=pca.transform(x_test)
```

# Apply Algorithms

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(multi_class='multinomial')
model.fit(x_train_transform,y_train)
```

```python
pre=model.predict(x_test_transfom)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pre)
```

# Check on query point

```python
query=np.array([[0.7,0.0]])
pre_prob=model.predict_proba(query)
pre_prob
```

```python
from mlxtend.plotting import plot_decision_regions
```
