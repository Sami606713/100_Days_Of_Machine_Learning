# Salaries

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
ax1.set_title("Histogram of Salaries")
sns.histplot(df['salary'],ax=ax1)

ax2.set_title("Kde of Salaries")
sns.kdeplot(df['salary'],ax=ax2)
plt.show()
```

```python
plt.figure(figsize=(15,7))
plt.title("Box plt of Salaries")
sns.boxplot(df['salary'])
plt.show()
```

```python
df['salary'].skew()
```

# Conclussion
1. we can see that data contain `Outliers`.
2. we can also see that data is `right skewed`.

# experience_level

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
ax1.set_title("Histogram of experience_level")
sns.barplot(x=df['experience_level'].value_counts().index,
            y=df['experience_level'].value_counts().values,ax=ax1)

ax2.set_title("Pie Chart of experience_level")
plt.pie(df['experience_level'].value_counts().values,
        labels=df['experience_level'].value_counts().values,autopct="%.2f")
plt.show()
```

# employment_type

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
ax1.set_title("Histogram of employment_type")
sns.barplot(x=df['employment_type'].value_counts().index,
            y=df['employment_type'].value_counts().values,ax=ax1)
ax1.tick_params(axis='x',rotation=90)

ax2.set_title("Pie Chart of employment_type")
plt.pie(df['employment_type'].value_counts().values,
        labels=df['employment_type'].value_counts().values,autopct="%.2f")
plt.show()
```

# job_title

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(15,7))
ax1.set_title("Histogram of job_title")
sns.barplot(x=df['job_title'].value_counts().nlargest(15).index,
            y=df['job_title'].value_counts().nlargest(15).values,ax=ax1)
ax1.tick_params(axis='x',rotation=90)

ax2.set_title("Pie Chart of job_title")
plt.pie(df['job_title'].value_counts().nlargest(15).values,
        labels=df['job_title'].value_counts().nlargest(15).values,autopct="%.2f")
plt.show()
```
